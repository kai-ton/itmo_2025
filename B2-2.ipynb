```python
# Случай без трения
T = 0.00001  # шаг времени, через который делаем рассчеты
Time = 10  # время втечение которого делаем расчеты
n = int(Time / T)  # кол-во шагов

m = 0.1
k1 = 5
k2 = 10
k = k1 + k2  # силы обеих пружин всегда будут сонаправлены

t = [i * T for i in range(n)]  # список всех моментов времени, в которые считаем
v = 0  # скорость груза
x = [0 for i in range(n)]  # список координат груза в наши моменты времени
x[0] = 0.01  # начальная координата
a = -k * x[0] / m  # начальное ускорение по второму закону Ньютона и формуле Гука
# делаем расчеты координаты, скорости и времени n - 1 раз
for i in range(1, n):
    x[i] = x[i - 1] + v * T + a * T**2 / 2
    v += a * T
    a = -k * x[i] / m
```

```python
import matplotlib.pyplot as plt
plt.plot(t, x);
```

```python
# Случай с трением
T = 0.00001
Time = 10
n = int(Time / T)

m = 0.1
k1 = 5
k2 = 10
k = k1 + k2
mu = 0.002  # коэффициент трения
g = 10  # ускорение свободного падения

t = [i * T for i in range(n)]
v = 0
x = [0 for i in range(n)]
x[0] = 0.01
a = -k * x[0] / m
for i in range(1, n):
    x[i] = x[i - 1] + v * T + a * T**2 / 2
    v += a * T
    # сила трения всегда направлена против вектора скорости
    if v > 0:
      a = -k * x[i] / m - mu * g
    else:
      a = -k * x[i] / m + mu * g
```

```python
import matplotlib.pyplot as plt
plt.plot(t, x);
```
